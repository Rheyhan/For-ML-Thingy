```python
#Why use shits like set then check each word when there's a module for it?

from sklearn.feature_extraction import text

corpus = ["To be, or not to be, that is the question:",
        "Whether 'tis nobler in the mind to suffer",
        "The slings and arrows of outrageous fortune,"]

#Fit Model
vectorizer = text.CountVectorizer()
vectorizer.fit(corpus)

#Print each vocabulary and its indice
print("Vocabulary: ", vectorizer.vocabulary_)

#print feature names
print(vectorizer.get_feature_names_out())
#Same thing but sorted from its indice
print(list(vectorizer.vocabulary_.keys()))

#Transform token
token_count = vectorizer.transform(corpus)
print(token_count)          #(list, indice)  occurence
```

```python
dense_tcm = token_count.toarray()
print(dense_tcm)
import pandas as pd

pd.DataFrame(dense_tcm,
             index=['corpus_0', 'corpus_1', 'corpus_2'],
             columns=vectorizer.get_feature_names_out())
```

```python
#searching for a specific word, E.G searching for the word "to" in each collumns

word="to"

for i in range(len(corpus)):
    print(f'"{corpus[i]}" :    {dense_tcm[i][vectorizer.vocabulary_[word]]}')
```

```python
#Extract token counts out of new text documents
txt = "That is the question and it is nobler in the mind."
vectorizer.transform([txt]).toarray()
print(vectorizer.transform([txt]).toarray(), end=2*"\n")
#Check the new feature names
print(vectorizer.get_feature_names_out())
print(vectorizer.vocabulary_)
```
